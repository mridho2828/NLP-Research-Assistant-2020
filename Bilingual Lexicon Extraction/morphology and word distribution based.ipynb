```python
# Implementasi model Rule Based Similarity

def rule_based(kata):
    temp = kata
    temp = re.sub(r'([a-z])age$', r'\1ase', temp)
    temp = re.sub(r'([a-z])ant$', r'\1an', temp)
    temp = re.sub(r'([a-z])archy$', r'\1arki', temp)
    temp = re.sub(r'([a-z])ary$', r'\1er', temp)
    temp = re.sub(r'([a-z])tion$', r'\1si', temp)
    temp = re.sub(r'([a-z])ique$', r'\1ik', temp)
    temp = re.sub(r'([a-z])ical$', r'\1is', temp)
    temp = re.sub(r'([a-z])ile$', r'\1il', temp)
    temp = re.sub(r'([a-z])ism$', r'\1isme', temp)
    temp = re.sub(r'([a-z])ist$', r'\1is', temp)
    temp = re.sub(r'([a-z])ive$', r'\1if', temp)
    temp = re.sub(r'([a-z])logue$', r'\1log', temp)
    temp = re.sub(r'([a-z])logy$', r'\1logi', temp)
    temp = re.sub(r'([a-z])ty$', r'\1tas', temp)
    temp = re.sub(r'([a-z])ure$', r'\1ur', temp)
    temp = re.sub(r'([a-z])ic', r'\1ik', temp)
#     if kata != temp:
    temp = re.sub(r'cch([a-df-hj-tv-z])', r'k\1', temp)
    temp = re.sub(r'ch([a-df-hj-tv-z])', r'k\1', temp)
    temp = re.sub(r'cc([b-df-hj-z])', r'k\1', temp)
    temp = re.sub(r'cc(e|i)', r'ks\1', temp)
    temp = re.sub(r'ck', r'k', temp)
    temp = re.sub(r'gh', r'g', temp)
    temp = re.sub(r'ou', r'u', temp)
    temp = re.sub(r'ph', r'f', temp)
    temp = re.sub(r'q', r'k', temp)
    temp = re.sub(r'rh', r'r', temp)
    temp = re.sub(r'sch([aeiou])', r'sk\1', temp)
    temp = re.sub(r'sc([a-df-hj-xz])', r'sk\1', temp)
    temp = re.sub(r'sc([eiy])', r's\1', temp)
    temp = re.sub(r'th', r't', temp)
    temp = re.sub(r'uu', r'u', temp)
    temp = re.sub(r'([a-z])x', r'\1ks', temp)
    temp = re.sub(r'xc([ei])', r'ks\1', temp)
    temp = re.sub(r'xc([a-df-hj-z])', r'ksk\1', temp)
    temp = re.sub(r'c([a-df-hj-xz])', r'k\1', temp)
    temp = re.sub(r'c(e|i|oe|y)', r's\1', temp)
    temp = re.sub(r'([b-df-hj-np-tv-z])\1', r'\1', temp)
    return temp
    
    return ''
```

```python
# Implementasi Spelling Similarity

from nltk.metrics.distance import edit_distance

def spelling(kata):
    hasil_spelling = []
    for word in set_inggris:
        skor_levenshtein = edit_distance(kata, word, transpositions=False)
        max_panjang = max(len(kata), len(word))
        skor = skor_levenshtein / max_panjang
        hasil_spelling.append((word, skor))
        
    hasil_spelling = sorted(hasil_spelling, key=lambda x: x[1])
    return hasil_spelling[:5]
```
